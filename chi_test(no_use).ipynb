```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

semesters = [
    "1021(01)", "1022(01)", "1031(01)", "1032(01)", "1041(01)", "1042(01)",
    "1051(01)", "1052(01)", "1061(01)", "1062(01)", "1071(01)", "1072(01)"
]

departments = [
    '企業管理學系'
]

for semester in semesters:
    # 讀取學期平均成績資料
    url_1 = f"https://raw.githubusercontent.com/JackyYang27/data-science-and-big-data/main/dataset/112-0004_%E6%A0%A1%E5%8B%99%E6%8B%8B%E8%BD%89%E8%B3%87%E6%96%99_%E5%AD%B8%E7%94%9F%E5%AD%B8%E6%9C%9F%E5%B9%B3%E5%9D%87%E6%88%90%E7%B8%BE_{semester}.csv"
    df_1 = pd.read_csv(url_1)
    df_filtered_1 = df_1[df_1['系所名稱'].isin(departments)]
    df_filtered_1['學期-班一般人數'] = pd.to_numeric(df_filtered_1['學期-班一般人數'], errors='coerce')
    df_filtered_1 = df_filtered_1.dropna(subset=['學期-班一般人數', '學期-班一般名次'])

    # 初始化 DataFrame
    top_students = pd.DataFrame()
    second_quintile_students = pd.DataFrame()

    # 分組並分別抓取前5%與5%-25%的學生
    for (dept, year_class), group in df_filtered_1.groupby(['系所名稱', '年級班別']):
        top_5_percent_index = max(int(round(len(group) * 0.05)), 1)
        top_25_percent_index = max(int(round(len(group) * 0.25)), 1)

        top_students_in_group = group.nsmallest(top_5_percent_index, '學期-班一般名次')
        second_quintile_students_in_group = group.nsmallest(top_25_percent_index, '學期-班一般名次').iloc[top_5_percent_index:]

        top_students = pd.concat([top_students, top_students_in_group], ignore_index=True)
        second_quintile_students = pd.concat([second_quintile_students, second_quintile_students_in_group], ignore_index=True)

    # 獲取前5%與5-25%學生的學號列表
    top_student_ids = top_students['學號'].tolist()
    second_quintile_student_ids = second_quintile_students['學號'].tolist()

    # 讀取選課資料
    url_2 = f"https://raw.githubusercontent.com/JackyYang27/data-science-and-big-data/main/dataset/112-0004_%E6%A0%A1%E5%8B%99%E6%8B%8B%E8%BD%89%E8%B3%87%E6%96%99_%E5%AD%B8%E7%94%9F%E9%81%B8%E8%AA%B2%E6%88%90%E7%B8%BE%E8%B3%87%E6%96%99_{semester}.csv"
    df_2 = pd.read_csv(url_2)


    # 篩選出學號對應的課程名稱和學分數，並且課程屬性名稱為選修或通識
    selected_courses_top = df_2[(df_2['學號'].isin(top_student_ids)) & (df_2['課程屬性名稱'].isin(['選修', '通識'])) & (~df_2['分發結果名稱'].isin(['人工退選','停修']))]
    selected_courses_second_quintile = df_2[(df_2['學號'].isin(second_quintile_student_ids)) & (df_2['課程屬性名稱'].isin(['選修', '通識'])) & (~df_2['分發結果名稱'].isin(['停修','人工退選']))]


    # 計算選課數量
    course_counts_top = selected_courses_top['課程名稱'].value_counts()
    course_counts_second_quintile = selected_courses_second_quintile['課程名稱'].value_counts()

    # 篩選出共同選課
    common_courses = course_counts_top.index.intersection(course_counts_second_quintile.index)
    course_counts_top_common = course_counts_top[common_courses]
    course_counts_second_quintile_common = course_counts_second_quintile[common_courses]

    # 計算選課比例
    total_top_students = len(top_student_ids)
    total_second_quintile_students = len(second_quintile_student_ids)

    course_proportions_top_common = course_counts_top_common / total_top_students
    course_proportions_second_quintile_common = course_counts_second_quintile_common / total_second_quintile_students

    # 建立列聯表
    data = {'Top 5%': course_counts_top_common, '5%-25%': course_counts_second_quintile_common}
    contingency_table = pd.DataFrame(data)

    # 進行卡方檢定
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # 輸出結果
    print(f"Chi-Square Value = {chi2}")
    print(f"P-Value = {p}")

    # 繪製柱狀圖
    fig, ax = plt.subplots(figsize=(12, 6))
    course_proportions_top_common.plot(kind='bar', color='blue', position=0, width=0.4, label='Top 5% Students')
    course_proportions_second_quintile_common.plot(kind='bar', color='red', position=1, width=0.4, label='5%-25% Students')

    plt.xlabel('課程名稱')
    plt.ylabel('選課比例')
    plt.title(f'前5%與5%-25%學生選課比例比較 - {semester}')
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 學期代碼列表
semesters = [
    "1021(01)", "1022(01)", "1031(01)", "1032(01)", "1041(01)", "1042(01)",
    "1051(01)", "1052(01)", "1061(01)", "1062(01)", "1071(01)", "1072(01)"
]

# 部門列表
departments = [
    '資訊管理學系'
]

# 遍歷每個學期進行分析
for semester in semesters:
    # 讀取學生平均成績數據
    url_1 = f"https://raw.githubusercontent.com/JackyYang27/data-science-and-big-data/main/dataset/112-0004_%E6%A0%A1%E5%8B%99%E6%8B%8B%E8%BD%89%E8%B3%87%E6%96%99_%E5%AD%B8%E7%94%9F%E5%AD%B8%E6%9C%9F%E5%B9%B3%E5%9D%87%E6%88%90%E7%B8%BE_{semester}.csv"
    df_1 = pd.read_csv(url_1)
    df_filtered_1 = df_1[df_1['系所名稱'].isin(departments)]
    df_filtered_1['學期-班一般人數'] = pd.to_numeric(df_filtered_1['學期-班一般人數'], errors='coerce')
    df_filtered_1 = df_filtered_1.dropna(subset=['學期-班一般人數', '學期-班一般名次'])

    # 初始化數據框架
    top_students = pd.DataFrame()
    last_percentile_students = pd.DataFrame()

    # 分組並識別前5%和後5%的學生
    for (dept, year_class), group in df_filtered_1.groupby(['系所名稱', '年級班別']):
        percentile_5 = max(int(round(len(group) * 0.05)), 1)
        top_students_in_group = group.nsmallest(percentile_5, '學期-班一般名次')
        last_students_in_group = group.nlargest(percentile_5, '學期-班一般名次')

        top_students = pd.concat([top_students, top_students_in_group], ignore_index=True)
        last_percentile_students = pd.concat([last_percentile_students, last_students_in_group], ignore_index=True)

    # 獲取前5%和後5%學生的學號列表
    top_student_ids = top_students['學號'].tolist()
    last_percentile_student_ids = last_percentile_students['學號'].tolist()

    # 讀取選課資料
    url_2 = f"https://raw.githubusercontent.com/JackyYang27/data-science-and-big-data/main/dataset/112-0004_%E6%A0%A1%E5%8B%99%E6%8B%8B%E8%BD%89%E8%B3%87%E6%96%99_%E5%AD%B8%E7%94%9F%E9%81%B8%E8%AA%B2%E6%88%90%E7%B8%BE%E8%B3%87%E6%96%99_{semester}.csv"
    df_2 = pd.read_csv(url_2)

    # 篩選出學號對應的課程名稱和學分數，並且課程屬性名稱為選修或通識
    selected_courses_top = df_2[(df_2['學號'].isin(top_student_ids)) & (df_2['課程屬性名稱'].isin(['選修', '通識'])) & (~df_2['分發結果名稱'].isin(['人工退選','停修']))]
    selected_courses_last = df_2[(df_2['學號'].isin(last_percentile_student_ids)) & (df_2['課程屬性名稱'].isin(['選修', '通識'])) & (~df_2['分發結果名稱'].isin(['停修','人工退選']))]

    # 計算選課數量
    course_counts_top = selected_courses_top['課程名稱'].value_counts()
    course_counts_last = selected_courses_last['課程名稱'].value_counts()

    # 篩選出共同選課
    common_courses = course_counts_top.index.intersection(course_counts_last.index)
    course_counts_top_common = course_counts_top[common_courses]
    course_counts_last_common = course_counts_last[common_courses]

    # 計算選課比例
    total_top_students = len(top_student_ids)
    total_last_students = len(last_percentile_student_ids)

    course_proportions_top_common = course_counts_top_common / total_top_students
    course_proportions_last_common = course_counts_last_common / total_last_students

    # 建立列聯表
    data = {'Top 5%': course_counts_top_common, 'Last 5%': course_counts_last_common}
    contingency_table = pd.DataFrame(data)

    # 進行卡方檢定
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # 輸出結果
    print(f"Chi-Square Value = {chi2}")
    print(f"P-Value = {p}")

    # 繪製柱狀圖
    fig, ax = plt.subplots(figsize=(12, 6))
    course_proportions_top_common.plot(kind='bar', color='blue', position=0, width=0.4, label='Top 5% Students')
    course_proportions_last_common.plot(kind='bar', color='orange', position=1, width=0.4, label='Last 5% Students')

    plt.xlabel('課程名稱')
    plt.ylabel('選課比例')
    plt.title(f'前5%與後5%學生選課比例比較 - {semester}')
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
